```python

def get_name():

    name = input ("Please enter your full name: ")
    get_name.name = name




def get_will():
    
    while True:
    
        will_status = input ("Do you have a valid Last Will and Testament in place?: ")
        get_will.will_status = will_status

        if will_status[0].lower() == "y":
            break

        elif will_status[0].lower() == "n":
            age = int(input("What is your age? "))
            get_will.age = age
            break

        else:
            print("\nOops! Invalid response. Please type 'yes' or 'no' and hit Enter")
            continue
    
            


def get_marital_status():
    
    while True:
    
        marital_status = input ("What is your current marital status?:\na) - Single \nb) - Married ICOP \nc) - Married ANC \nd) - Married ANC with Accrual \ne) - Divorced \nf) - Widowed\n ")
        get_marital_status.marital_status = marital_status
        valid_response = ("a", "b", "c", "d", "e", "f")
        
        if marital_status in valid_response:
            break
        else:
            print ("\nOops! Invalid response. Please enter the relevant letter from the options above and hit Enter")



    
def get_invest_term():
    
    while True:
    
        invest_term = input ("Would you like to invest monthly or as a lump sum? ")
        get_invest_term.invest_term = invest_term

        if invest_term[0].lower() == "m":
            invest_amount_monthly = int(input("How much would you like to invest per month? R: "))
            get_invest_term.invest_amount_monthly = invest_amount_monthly
            break

        if invest_term[0].lower() == "l":
            invest_amount_lump = int(input("How big of a lump sum would you like to invest? R: "))
            get_invest_term.invest_amount_lump = invest_amount_lump
            break
            
        else:
            print ("\nOops! Invalid response. Please type 'monthly' or 'lump sum' and hit Enter")
            continue
        
    
        
        
def get_objectives():
    
    
    while True:
        
        objectives = (input("What are your investment objectives? \na) - Capital Growth \nb) - Income \nc) - Growth and Income\n"))
        get_objectives.objectives = objectives
        valid_response = ("a", "b", "c")
    
        if objectives == "a":
            investment_horizon = input ("How long do you plan to invest for? \na) - Less than 3 years \nb) - 3 to 5 years \nc) - 5 to 10 years \nd) - 10+ years\n")
            growth_expectations = input ("What type of return do you expect from your investment? \na) inflation (roughly what you would earn with a bank deposit) \nb) inflation +2% \nc) inflation +3% \nd) inflation +4% \ne) inflation +5% \nf) inflation +6% \ng) inflation +7%\n ")
            get_objectives.investment_horizon = investment_horizon
            get_objectives.growth_expectations = growth_expectations
            break

        elif objectives == "b":
            income_amount = input("How much income do you need from your investment? ")
            income_growth_expectations = input ("What type of return do you expect from your investment? \na) inflation (roughly what you would earn with a bank deposit) \nb) inflation +2% \nc) inflation +3%\n")
            get_objectives.income_amount = income_amount
            get_objectives.income_growth_expectations = income_growth_expectations
            break

        elif objectives == "c":
            income_amount = input("How much income do you need from your investment? ")
            investment_horizon = input ("How long do you plan to invest for? \na) - Less than 3 years \nb) - 3 to 5 years \nc) - 5 to 10 years \nd) - 10+ years\n")
            growth_expectations = input ("What type of return do you expect from your investment? \na) inflation (roughly what you would earn with a bank deposit) \nb) inflation +2% \nc) inflation +3%\n ")
            get_objectives.income_amount = income_amount
            get_objectives.investment_horizon = investment_horizon
            get_objectives.growth_expectations = growth_expectations
            break
        
        else:
            print ("\nOops! Invalid response. Please enter 'a', 'b', or 'c'")
            continue
    


       
def get_risk_profile():
    
    risk_profile = 0
    
    
    compound_understanding = int(input("How would you rate you understanding of the following on a 1 - 5 scale: \n\nThe effects of compound interests "))
    risk_profile += compound_understanding
    
    negative_returns_understanding = int(input("\nThe effect of negative returns? "))
    risk_profile += negative_returns_understanding

    risk_understanding = int(input("\nThe effect of investment risk and market volatility? "))
    risk_profile += risk_understanding

    risk_appetite = int(input("\nOn a scale from 1 - 5, how would you rate your tolerance for risk? "))
    risk_profile += risk_appetite
    
    risk_profile = risk_profile/4
    get_risk_profile.risk_profile = risk_profile

    
    
def advice():
    
    print("")
        
    if get_will.will_status[0].lower() == "n" and get_will.age > 35:
        print ("\nYou said in the quiz that you don't have a valid will and testamony. You really should start thinking about putting one\ntogether to make sure your loved ones are taken care of and to make sure that there are processes in order that enable them\nto greive properly without the stress of sorting out your estate after your passing")

    if get_will.will_status[0].lower() == "y":
        print ("\nThe fact that you have a valid will in place already puts your loved ones in a much safer position than they would be\notherwise. It also puts you far ahead of the average South African citizen in terms of your financial risk management.\nGood job!")
    
    if get_marital_status.marital_status == "b":
        print ("\nI see that you are married in community of property. Did you know that you can still enter into an ante-nuptual contract with\nyour partner? Though this can be quite a tedious process, your assets will be much better protected in the case of death or\ndivorce. However, one advantage of remaining married in community of property is that you can save yourself quite a bit on\ncapital gains taxes. It might also interest you to know that inheritences can be excluded from community of property marriages as long as it is in the will of the decease from whom the asset is inherited")
    
    if get_invest_term.invest_term[0].lower() == "m" and get_invest_term.invest_amount_monthly < 500:
        print ("\nYou stated that you plan on investing R{} per month. You should know that most of the platforms we use to help our clients\ninvest their money require a minimum of R500 per month".format(get_invest_term.invest_amount_monthly))

    if get_invest_term.invest_term[0].lower() == "l" and get_invest_term.invest_amount_lump < 20000:
        print ("\nYou stated that you would like to invest a lump sum of R{}. Please note that most of the platforms we use to help our\nclients invest their money require a minimum lump sum investment of R20 000".format(get_invest_term.invest_amount_lump))

    if get_objectives.objectives == "a" and ((get_objectives.investment_horizon == "a" or get_objectives.investment_horizon == "b") and (get_objectives.growth_expectations > "d")):
        print("\nYou mentioned that your primary investment objective is capital growth but that you plan on investing for less than 3 years. Please note that portfolios structured for growth are quite volatile and could yield dissapointing results after such a short investment term. It is recommended that you either aim for lower safer returns or extend your investment horizon to exceed at least 5 years.")
        
    if get_risk_profile.risk_profile <= 3:
        print ("\nBased on your understanding of some key investment principles as well as your tolerance for risk, you have been rated with a\nrisk profile of {} This is a fairly low risk profile, better suited to short term investments and investments that emphasize\ncapital preservation".format(get_risk_profile.risk_profile))

    if get_risk_profile.risk_profile > 3:
        print ("\nBased on your understanding of some key investment principles as well as your tolerance for risk, you have been rated with a\nrisk profile of {} This is a fairly high risk profile, better suited to long term and aggressive investing.".format(get_risk_profile.risk_profile))
    
    
    
    
    
```
